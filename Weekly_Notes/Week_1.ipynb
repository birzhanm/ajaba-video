#Introduction to Working with Videos and OpenCV library
A video is just a sequence of images and an accompanying audio. The images which constitute video are called frames. The number of frames in a given second of video is called *frames per second (fps)*. One of the basic ways of working with videos invlove splitting a video into frames, editing these frames and re-combining the frames into a resuting video. There are many tools and libraries fot for this kind of operations. We can choose a popular library called OpenCV, which stands for Open Computer Vision. Let us try to use this library to split a video into frames and then combining the frames to make a video. 

```
# we need to install opencv first
!pip install numpy==1.15.0
!pip install imgaug==0.2.8
!apt-get -qq install -y libsm6 libxext6 && pip install -q -U opencv-python
import cv2
```

```
!pip3 install wget
import wget
sample_video_url = 'https://sample-videos.com/video123/mp4/360/big_buck_bunny_360p_2mb.mp4'
wget.download(sample_video_url, 'sample_video.mp4')
```

```
video_path = 'sample_video.mp4'
output_dir = 'frames'

import os
# video to frames function
def vid2frames(video_path, output_dir):
  vid = cv2.VideoCapture(video_path)
  success, image = vid.read()
  count = 0
  while success:
    frame_path = os.path.join(output_dir, "{}_frame.jpg".format(count))
    cv2.imwrite(frame_path, image)
    success, image = vid.read()
    print('Read a new frame: ', success)
    count += 1
    

# start a directory for frames, if there is no such a directory
if not os.path.exists(output_dir):
  os.makedirs(output_dir) 
# run the above function
vid2frames(video_path, output_dir)
```

* PyTorch allows you to skip the nitty-gritty details of training neural network. Once you define a neural network and supply the data, PyTorch will take care of training your model itself.
* PyTorch allows to parallerlize training process by introducing GPU (Graphical Processing Unit) support.
